## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Create a copy for wrangling
df_wrangled = df.copy()

print("DataFrame after creating a copy:")
display(df_wrangled.head())
````

````python
# Drop 'ID', 'Customer_ID', and 'SSN' as they are identifiers and likely not useful features
# 'Name' is also dropped due to high cardinality and likely no predictive power
df_wrangled = df_wrangled.drop(['ID', 'Customer_ID', 'SSN', 'Name'], axis=1)

print("DataFrame after dropping identifier columns:")
display(df_wrangled.head())
````

````python
# Handle 'Credit_Score' (Target Variable) - Ordinal Mapping
# We map Poor to 0, Standard to 1, and Good to 2 for potential modeling purposes.
credit_score_mapping = {'Poor': 0, 'Standard': 1, 'Good': 2}
df_wrangled['Credit_Score'] = df_wrangled['Credit_Score'].map(credit_score_mapping)

print("DataFrame after mapping 'Credit_Score':")
display(df_wrangled.head())
````

````python
# Handle 'Credit_Mix' - Ordinal Mapping
# We map Bad to 0, Standard to 1, and Good to 2 as this is also an ordered category.
credit_mix_mapping = {'Bad': 0, 'Standard': 1, 'Good': 2}
df_wrangled['Credit_Mix'] = df_wrangled['Credit_Mix'].map(credit_mix_mapping)

print("DataFrame after mapping 'Credit_Mix':")
display(df_wrangled.head())
````

````python
# Handle 'Payment_of_Min_Amount' - Convert 'NM' to NaN and then apply Binary Mapping
# We map 'No' to 0 and 'Yes' to 1 as this is a binary category.

# First, replace 'NM' with NaN
df_wrangled['Payment_of_Min_Amount'] = df_wrangled['Payment_of_Min_Amount'].replace('NM', np.nan)

# Then, apply the binary mapping
payment_min_mapping = {'No': 0, 'Yes': 1}
df_wrangled['Payment_of_Min_Amount'] = df_wrangled['Payment_of_Min_Amount'].map(payment_min_mapping)

print("DataFrame after handling 'NM' and mapping 'Payment_of_Min_Amount':")
display(df_wrangled.head())

# to check for missing values again after this step
print("\nMissing values in 'Payment_of_Min_Amount' after handling 'NM':")
print(df_wrangled['Payment_of_Min_Amount'].isnull().sum())
````

````python
# Handle 'Type_of_Loan' - Splitting and creating binary columns
if 'Type_of_Loan' in df_wrangled.columns:
    print("Processing 'Type_of_Loan'...")

    # Clean up the strings in the 'Type_of_Loan' column
    # Remove 'and' and any surrounding spaces, case-insensitive
    df_wrangled['Type_of_Loan_cleaned'] = df_wrangled['Type_of_Loan'].astype(str).str.replace(r'\s*and\s*', ',', regex=True)
    # Replace multiple commas with a single comma
    df_wrangled['Type_of_Loan_cleaned'] = df_wrangled['Type_of_Loan_cleaned'].str.replace(r',+', ',', regex=True)
     # Remove leading/trailing commas that might result from the replacement
    df_wrangled['Type_of_Loan_cleaned'] = df_wrangled['Type_of_Loan_cleaned'].str.strip(',')


    # Step 1: Identify all unique loan types across all entries using the cleaned column
    all_loan_types = set()
    for loan_list_str in df_wrangled['Type_of_Loan_cleaned'].dropna(): # Handle potential NaNs
        # Split by comma and strip whitespace, then add to the set
        types = [loan.strip() for loan in loan_list_str.split(',') if loan.strip() != ''] # Added check for empty strings after split
        all_loan_types.update(types)

    # Remove potential "nan" string if it exists due to NaN handling
    if 'nan' in all_loan_types:
        all_loan_types.remove('nan')

    print(f"Found {len(all_loan_types)} unique loan types after cleaning.")
    print(f"Unique loan types found: {list(all_loan_types)}") # Print the actual types


    # Step 2: Create new binary columns for each unique loan type
    for loan_type in all_loan_types:
        # Create a new column for each loan type. 1 if the type is present, 0 otherwise.
        df_wrangled[f'Has_{loan_type.replace(" ", "_")}'] = df_wrangled['Type_of_Loan_cleaned'].apply(
            lambda x: 1 if isinstance(x, str) and loan_type in [t.strip() for t in x.split(',') if t.strip() != ''] else 0
        )

    # Step 3: Drop the original and cleaned 'Type_of_Loan' columns
    df_wrangled = df_wrangled.drop(['Type_of_Loan', 'Type_of_Loan_cleaned'], axis=1)

    print("DataFrame after handling and cleaning 'Type_of_Loan':")
    display(df_wrangled.head(10))
else:
    print("'Type_of_Loan' column not found, skipping loan type handling.")
````

````python
# Apply One-Hot Encoding to the remaining nominal columns (Occupation, Payment_Behaviour)
nominal_cols = ['Occupation', 'Payment_Behaviour']
# Ensure the columns actually exist before attempting to one-hot encode
nominal_cols_to_encode = [col for col in nominal_cols if col in df_wrangled.columns]

if nominal_cols_to_encode:
    print(f"Applying One-Hot Encoding to: {nominal_cols_to_encode}")
    df_wrangled = pd.get_dummies(df_wrangled, columns=nominal_cols_to_encode, drop_first=True)

    print("DataFrame after One-Hot Encoding nominal variables:")
    display(df_wrangled.head())
    print(f"New number of columns: {df_wrangled.shape[1]}")
else:
    print("No nominal columns found for One-Hot Encoding.")
````

````python
# Final check of the wrangled dataframe
print("\nFinal Wrangled DataFrame Information:")
df_wrangled.info()
print("\nFinal Wrangled DataFrame Description:")
display(df_wrangled.describe())
````

````python
int_columns = [
    'Num_Bank_Accounts', 'Num_Credit_Card', 'Num_of_Loan',
    'Num_of_Delayed_Payment', 'Num_Credit_Inquiries', 'Credit_History_Age'
]

# Convert to integers safely
for col in int_columns:
    if (df_wrangled[col] % 1 == 0).all():  # check if all values are whole numbers
        df_wrangled[col] = df_wrangled[col].astype('int64')
df_wrangled.dtypes
````

#### Chart - 12

````python
# Chart - 12: Impact of Occupation on Credit Score (Bivariate - Categorical vs. Categorical)
# Since Occupation is one-hot encoded, we can visualize the average credit score
# for each occupation category compared to the overall average.

# Identify one-hot encoded occupation columns
occupation_cols = [col for col in df_wrangled.columns if col.startswith('Occupation_')]

if occupation_cols:
    print("Visualizing average Credit Score per Occupation...")

    # Calculate the average credit score for each occupation
    # We group by the one-hot encoded columns where the value is 1 (meaning the customer has that occupation)
    # This requires reshaping or iterating. A simpler way is to calculate the mean of Credit_Score
    # for rows where the specific occupation column is 1.

    occupation_avg_credit = {}
    for occ_col in occupation_cols:
        # Calculate the mean Credit_Score for rows where this occupation column is 1
        # We also check if there are any such rows to avoid division by zero or errors on empty slices
        if df_wrangled[occ_col].sum() > 0:
             avg_score = df_wrangled.loc[df_wrangled[occ_col] == 1, 'Credit_Score'].mean()
             # Get the original occupation name from the column name
             original_occ_name = occ_col.replace('Occupation_', '')
             occupation_avg_credit[original_occ_name] = avg_score

    # Convert to DataFrame for easier plotting
    occupation_avg_credit_df = pd.DataFrame(list(occupation_avg_credit.items()),
                                           columns=['Occupation', 'Average_Credit_Score'])

    # Sort by average credit score for better visualization
    occupation_avg_credit_df = occupation_avg_credit_df.sort_values('Average_Credit_Score')

    plt.figure(figsize=(14, 8))
    sns.barplot(data=occupation_avg_credit_df, x='Average_Credit_Score', y='Occupation', palette='viridis')
    plt.title('Average Credit Score by Occupation')
    plt.xlabel('Average Credit Score (0: Poor, 1: Standard, 2: Good)')
    plt.ylabel('Occupation')
    plt.xlim(0, 2) # Set x-axis limits to reflect the credit score scale
    plt.show()

else:
    print("No one-hot encoded Occupation columns found for plotting.")
````

##### 1. Why did you pick the specific chart?

A bar plot is effective for comparing a single numerical value (average credit score)
across multiple discrete categories (occupations). Sorting the bars makes it easy
to see which occupations are associated with higher or lower average credit scores.


##### 2. What is/are the insight(s) found from the chart?

This chart will show which occupations tend to have higher or lower average credit scores.
This can help identify occupation groups that are potentially higher or lower risk.

##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.
